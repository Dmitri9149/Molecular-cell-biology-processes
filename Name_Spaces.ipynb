## Name Spaces 

### Cell Circulation

We may split our body to two parts: 'solid' body and circulating, mobile  cells like erithrocytes, lymphosytes, antigenpresenting cells (macrophages, dendritic cells, ..) etc..
For the circulating cells the 'solid' body work as computational environment, as stratified adress space. To understand immune interactions, we have to understand how the immune cells circulate.  
The 'solid' body is composed of cells too, of course. The 'gas' of circulating cells is inclued in the 'solid' body environment.   
In this part we will concentrate how the environment determine (or better to say : may determine, because it is our reconstruction of biological events) the circulation of the mobile cells. We take 'lymphocyte' as example. 

We will use two dimentional manifold to model the adress space and lymphocyte circulation. The model is simple: the surface of the sphere is 'blood' , in the blood are 'swimming' different organs. The meaning is : we can travel from any point in 'blood' to border of any organ. The handles correspond to lymph collectors which drainage organs (lymph nodes reside on the collector path) and come back to blood again.

 
In the drawing above we may see the hypothetical topology of human (or an animal) body name space. It is very simplified picture, but there are the main ingredients to describe some important space..time events from lymphocyte point of view. We will use pi-calculus for the system description.

We think about our body as 2-D surface partitioned into unintersecting area. The broders (circles) are the part of the partitioning. 
We start from the Blood.  
The Blood area is like a 'bus' in computer motherboard. All 'organs':  Lung, Skin, Liver, etc.. are 'swimming' in the Blood and a lymphocite or erytrocyte may 'swim' to any organ from the Blood. 
We will use only Skin and Lung, it is enought for us.  
We may see the 'handle' which is starting at an organ at one side and attached to the Blood area at another side. The handle represent Lymph Collectors (on the organ side) and lymph vessels coming back to the Blood via veins on the other side. The lymph collectors are a garbage colectors filtrating our organs.  
And this is not the only function of the system. Actually, the drainage , garbage lymph collectors work as 'Data Lakes' which collect raw data. The data are 'antigens' in raw form. Special cells: antigen presenting cells process the raw data and tramsform it into refined form ('data warehouse' form). This refined form is 'understandable' for the lymphocytes. All the interactions take place in Lymph Nodes.  
In the middle of a handle we may see the Lymph Nodes: the place where immune cells interact between each other. The Lymph Nodes are a part of the filtration drainage system.  

The topology is not trivial: our partitioning is not a tree or a forest, where we have mainly inclusion of subparts into other subparts, but correspond to a cyclic graph. We may go from Blood to Skin, then to Lymph Collector, Lymph Node , Lymph Vessels and to Blood again by crossing each border between subparts only one time. 

The borders are very important part of our system. In our body a border is a collection of cells too. To come to an organ lymphocyte has to interact with border cells to get the permission to come in. Such molecules as integrins are responsible for the process to come from blood to organs for lymphocytes. We would like to stress the role of 'normal' structural environment in immune response. Cytotoxic lymphocytes, the key component of antivial responce, have to have normall physical access to infected cell. Moreover, the infected cell is to be 'mostly' normal at this time.  Lymphocyte just send a signal for the infected cell to make 'suicide'. Which is actually nothing more as a 'gracefull shutdown' of the cell process (like we safelly 'Shut down' our operating system).





### Processes on the organs boarders.
The cells of the borders are named in the way:    
$$bl$$ Where $b$ is for Blood and $l$ for Lung.  
$bl^+$ correspond to transition from Blood to Lung.  
$bl^-$ from Lung to Blood. 
The + and - correspond to two possible orientations of out circles-borders and posibility to come in or to come out of an area. 

Vocabulary:  
$bl$ : the boarders between Blood and Lungs;  
$bs$ : Blood and Skin;  
$oc$ : between organs and Lymph Collector  
(we skip dependence from an  organ like SC or LC, it will be OC for all organs).  
$cn$ : Collector / Lymph Node  
$nv$ : Lymph Node / Lymph Vessel   
$vb$ : Lymph Vessel / Blood
(same comment as above).

We may study algebraic topology by triangulating our objects or by gluing circles of different dimention. Here we 'glue circles'.

### Process calculus
For the description of our system we will use the Milnor's pi-calculus. 
We have the set of processes corresponding to the borders: 
$$Borders = sc^+!|sc^-!|bl^+!|bs^-!|bs^-!|bs^+!|oc^+!|oc^-!|cn^+!|cn^-!|cv^+!|cv^-!|vb^+!|vb^-!$$  

We will describe a lymphocyte which demonstrate a 'homing': it permanently moves from Blood to Lungs, then to Lymph Collector, to Lymph Node, to Lymph Vessels and finally to Blood. Then in moves to Lungs again: it monitors the Lungs.  
The behaviour is given by a recursive equation: 
$$Lymphocyte = \overline{bl^+}.\overline{oc^+}.\overline{cn^+}.\overline{nv^+}.\overline{vb^+}.Lymphocyte$$

And the total system is given by the expression:
$$Borders|Lymphocyte$$ 
where we compose in parallel the two subparts.
