### Cytotoxic and Delayed Type Hypersensitivity

3.It is my personal opinion basing of my intuition:   
The main factor in coronavirus protection is T cell cytotoxic response. Cytotoxic cell find recently infected cell and send signal for 'suicide' to the cell.  Actually it is signal to stop the cell processes in 'gracefull' way: all the job to stop processes is on the INFECTED CELL side.  
This kind of response is possible ONLY IF: the infected cells still reply to the signals, the microstructure of infected tissue still allow for the T lymphocytes to reach the infected cells.  
If infection continues, the Delayed Type Hypersensitivity (DTH)  T cell come into play. It is very different responce: the cells marked infected part of a tissue and neutrophils like cells penetrate the part: the neutrophils necrotize 'in gracefull way' the PART OF TISSUE.  
The high temperature at the beginning of infection is counterpart of the t cell cytotoxic type of the immune responce. It is much more 'gentle' than the DTH response.  
The early decrease in temperature increase chances the T cell cytotoxic responce will not be enough for the recovery from virus and DTH will come into play. Temperature is very hight load to all our organs , but late complications because of virus intoxication and DTH responce is very high price for our organs too.

### Virus survival. 'Train in Metro'

The citation is from https://virologyj.biomedcentral.com/articles/10.1186/s12985-020-01418-7

"With initial viral loads broadly equivalent to the highest titres excreted by infectious patients, viable virus was isolated for up to 28 days at 20 °C from common surfaces such as glass, stainless steel and both paper and polymer banknotes. Conversely, infectious virus survived less than 24 h at 40 °C on some surfaces."



4.As we can see the coronavirus demonstrate very high survival. In case of 0 or minus temperature, the 28 days may become several months. It is very different from influenza virus: "The influenza virus that causes the flu can live for up to three days on hard surfaces like subway poles and door handles."(from https://www.insider.com/how-long-do-flu-germs-live#:~:text=The%20influenza%20virus%20that%20causes,5%20minutes%20on%20the%20body.)
